```python
import geopandas as gpd
import fiona
```

```python
gdf_all = gpd.read_file('../Data/cbs/wijkenbuurten_2023_clean.gpkg')
gdf_amr = gpd.read_file('../Data/cbs/arbeidsmarktregios.gpkg')
```

```python
allcols = gdf_all.columns.tolist()
allcols.insert(6, 'arbeidsmarktregio')
```

```python
def get_amregio(geometry):
    # select the first matching region
    maxarea = 0
    amr = None
    for index, row in gdf_amr.iterrows():
        if row.geometry.contains(geometry):
            return row.statnaam
        elif row.geometry.intersects(geometry):
            if row.geometry.intersection(geometry).area > maxarea:
                maxarea = row.geometry.intersection(geometry).area
                amr = row.statnaam
    return amr
```

```python
sample = gdf_all.sample(5)
sample['arbeidsmarktregio'] = sample.geometry.apply(get_amregio)
sample = sample[allcols]
sample
```

```python
gdf_all['arbeidsmarktregio'] = gdf_all.geometry.apply(get_amregio)
gdf_all = gdf_all[allcols]
gdf_all.head(3)
```

```python
gdf_all.to_file('../Data/cbs/wijkenbuurten_2023_clean.gpkg', driver='GPKG')
```
