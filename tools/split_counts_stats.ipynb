```python
import geopandas as gpd
import re

bu = gpd.read_parquet('data/gemeenten_concatenated_counted.parquet')
wk = gpd.read_parquet('data/wijken_concatenated.parquet')
bu = gpd.read_parquet('data/buurten_concatenated.parquet')
bu.head(3)
```

```python
bu.columns.values

# regex with Lx_x_count
regex = re.compile(r'L\d{1}_\d{1}_count')
# get columns that match the regex
columns = [col for col in bu.columns if regex.match(col)]
countcols = columns

counts = bu[['gemeentecode'] + countcols]
# stats is without the count columns
stats = bu.drop(columns=countcols)
stats.head(5)
```

```python
counts.head(5)
```

```python
scale = "buurten"

stats.to_parquet(f"data/{scale}/{scale}_stats.parquet")
counts.to_parquet(f"data/{scale}/{scale}_counts.parquet")
```
