```python
import geopandas as gpd

import gc
filters = ["filter0", "filter1", "filter2"]
scales = ["buurten", "wijken"]
gemeenten_selectie = ["Eindhoven", "Amsterdam", "Rotterdam", "Nijmegen", "Maastricht"]

cols = ['L0_shannon', 'L1_shannon', 'L0_altieri', 'L1_altieri', 'L0_leibovici', 'L1_leibovici']
```

```python
for filter in filters:
    for scale in scales:
        for gm in gemeenten_selectie:
            # normalise the columns
            df = gpd.read_parquet(f"results_normalised/{filter}/{scale}/{gm}_{scale}.parquet")
            for col in cols:
                # apply min-max normalisation
                df[f"{col}_n"] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())
            df.to_parquet(f"results_normalised/{filter}/{scale}/{gm}_{scale}.parquet")
            del df
            gc.collect()
   
```
