```python
from sklearn.datasets import make_blobs
from spatialentropy import altieri_entropy, leibovici_entropy
import matplotlib.pyplot as plt



entropies_a = []
entropies_l = []
fig, ax = plt.subplots(10,1, figsize=(10, 50))
# increment in steps of 0.1
for i in range(0, 10):
    j = i
    X, y = make_blobs(n_samples=200, n_features=2, centers=3, cluster_std=j, random_state=42)
    entropies_a.append(leibovici_entropy(X, y).entropy)
    
    # add a plot for each value of j
    ax[i].scatter(X[:,0], X[:,1], c=y)

entropies_a
```

```python
fig, ax = plt.subplots() 
ax.plot(entropies_a, label='Altieri Entropy')
ax.plot(entropies_l, label='Leibovici Entropy')
ax.set_xlabel('Cluster Standard Deviation')
ax.set_ylabel('Entropy')
ax.legend()
```
