```python
import geopandas as gpd
import numpy as np

wijken = gpd.read_parquet('data/wijken_stedent_counts.parquet')
len(wijken)
```

```python
wijken.head(3)
```

```python
# split into 6 parts
wijken = np.array_split(wijken, 8)

for i, g in enumerate(wijken):
    g.to_parquet(f'data/wijken_parts/wijken_{i}.parquet')
    print(f"wrote part {i} with {len(g)} rows to data/wijken_parts/wijken_{i}.parquet")
```
