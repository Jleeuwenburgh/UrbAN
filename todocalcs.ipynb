# Gemeenten

```python
gemeenten = gpd.read_parquet('data/gemeenten_concatenated.parquet')

gemeenten.head(5)
```

```python
cols1 = gpd.read_parquet('data/gemeenten_concatenated.parquet').columns.values
cols2 = gpd.read_parquet('data/gemeenten_RE.parquet').columns.values

set(cols1) - set(cols2)
```

### Aanwezig
* Shannon entropy on all filter levels
* Relative entropy (shannon)

### Niet aanwezig
* Altieri & Leibovici entropies
* Liveliness
* Autocorrelation
